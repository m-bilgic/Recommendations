### (a) Split movie Genre into 0 or 1

```python
#replace the index with simple sequential numbers
data1 = pd.DataFrame(groups_rm[190]).reset_index()
genre_iter = (set(x.split('|')) for x in data1.Genres)
genre_iter
movie_genres = sorted(set.union(*genre_iter))
```

```python
dummies = pd.DataFrame(np.zeros((len(data1), len(movie_genres))), columns=movie_genres)
```

```python
for i, gen in enumerate(data1.Genres):
        dummies.ix[i, gen.split('|')] = 1
```

```python
movies_table = data1.join(dummies)
movies_table
```

### (c) Compute mean ratings

```python
#Number of distinct movies in the rating data
distinct_movie_rating = rating['MovieID'].value_counts()
dmf=pd.DataFrame({'Distinct': distinct_movie_rating})
dmf[dmf.index ==231]
```

```python
#The average rating for each movie
grp = tmp_rm.groupby('MovieID')
mean_rating = grp.mean()
mean_rating[mean_rating.index == 231].Rating
#mean_rating[mean_rating.MovieID == 231].Rating
```

```python
ratingname= ['meanrating']
dummies1 = pd.DataFrame(np.zeros((len(data1), 1)), columns = ratingname)

```

```python
j = 0
for i in movies_table.MovieID:
    if j <= len(dummies1):
        a = dmf[dmf.index == i].Distinct.values #number of ratings for this movie
        b = mean_rating[mean_rating.index == i].Rating.values #all uers mean rating for this moive
        c= movies_table[movies_table.MovieID == i].Rating.values #the rating of this movie that the user give
        d = dmf[dmf.index == i].Distinct.values -1 #number of ratings for this movie except this user
        meanrating = (a*b-c)/d
    
        dummies1.ix[j]= meanrating
        j = j+1
#print dummies1   
```

```python
tmp_table = movies_table.join(dummies1)
tmp_table
```
