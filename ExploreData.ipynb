# Understanding of Movielens Data

## Import packages

```python
%matplotlib inline
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

import pandas as pd
```

## Load the data

```python
#read data from file 
names1 = ['MovieID', 'Title',  'Genres']
data1 = pd.read_csv("/Users/leslieniu/Desktop/ml-10M100K/movies.dat", sep='::', names=names1)


names2 = ['UserID', 'MovieID', 'Rating', '']
data2 = pd.read_csv("/Users/leslieniu/Desktop/ml-10M100K/ratings.dat", sep='::', names=names2,usecols=range(3))

names3 =['UserID', 'MovieID', 'Tag','']
data3 = pd.read_csv("/Users/leslieniu/Desktop/ml-10M100K/tags.dat", sep='::', names=names3,usecols=range(3))

#merge the data 
data = pd.merge(pd.merge(data1, data2),data3)
print "Number of rows: %i. Number of columns: %i" %(data.shape[0],data.shape[1])
data.head()  # print the first 5 rows
```

```python
data['Title'] = [t[0:-6] for t in data.Title]
```

```python
data.Tag
```

## Split up the genres

```python
#split the genres
#determine the unique genres
genresset = set()

for m in data.Genres:
    genresset.update(g for g in m.split('|'))
genresset = sorted(genresset)

#make a column for each genre
for genre in genresset:
    data[genre] = [genre in movie.split('|') for movie in data.Genres]
```

```python
data.iloc[0:10,0:25]
```

```python
data['Title'] = [t[0:-6] for t in data.Title]
data.head()
```

## Explore the Data

### 1.how many movies in each genres

#### count 

```python
#sum sums over rows by default
genre_count = np.sort(data[genresset].sum())[::-1]
pd.DataFrame({'Genre Count': genre_count})
```

#### plot the histogrm

```python
pd.DataFrame({'Genre Count': genre_count}).plot(kind='bar', figsize=(12, 6), title='Genres by film count')
plt.show()
```

### 2. On average, how many genres does a movie have

```python
#axis=1 sums over columns instead
genre_count = data[genresset].sum(axis=1) 
print "Average movie has %0.2f genres" % genre_count.mean()
genre_count.describe()
```
